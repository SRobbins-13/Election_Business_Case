# Directing Political Ad Spending Business Use Case - MVP
Samuel Robbins

### Opportunity

I am a data scientist for a progressive Super PAC/ad company that is looking to determine which house districts would benefit from concentrated ad buys to flip seats and hold on to seats in competitive house districts. With political ad spending at an [all time high](https://www.forbes.com/sites/howardhomonoff/2020/12/08/2020-political-ad-spending-exploded-did-it-work/), and voters being inundated with more information than they know what to do with, targeted ad spending has the potential to play a decisive roll in future federal elections. The guiding question of this project is which house districts have the best potential to reach and convert voters for a reasonable cost and return on investment (ROI). 

### Impact Hypothesis

The collection and analysis of US voter data combined with the reach/cost of the major US media markets will allow progressive Super PACs and ad makers to more effectively allocate financial resources towards the most competitive house races with the biggest ROI. 

### Solution Path

As a first step, I will determine the partisan lean of each house district based on vote totals. This will involve creating functions to determine vote percentages based on raw vote totals. On a second iteration, I will examine the electoral history of the most competitive districts to determine the overall partisan trend and trajectory of those districts. 

By linking the house district maps to partisan lean and media market size, this will allow ad makers/buyers to more efficiently allocate finite resources to competitive districts. 

### Data

Election data provided by [MIT Election Labs](https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/IG0UN2) on the results of US House races from 1976 to the present. This will allow me to analyze not only the most recent house results, but also observe partisan trends over time.

Media market reach data provided by [Media Tracks](https://mediatracks.com/resources/nielsen-dma-rankings-2021/). This dataset will allow me to link each house district to it's closest media market and the reach of that market (defined as the number of homes in that market).

### Methodology

I used Google Sheets for data cleaning and prelimary data analysis and exploration. 

Based on raw vote totals I determined the percent of the vote share that went to the two major parties (Democrats and Republicans) and used a pivot table grouping to engineer the 'partisan lean' feature for each house district. In doing so, I have identified the 41 most competitive house districts (see below) - defined as those districts with a partisan margin of less than 5 points on either the R or D side.

I will link each major media market with the zipcodes in that market and link the zipcodes to the house districts. This will allow me to identify the house districts in the markets with the most reach.

I will use tableau for data visualization to create a dashboard that shows the US house district map with characterizations for partisan lean and market reach.
